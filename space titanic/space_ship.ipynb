```python
df_train = df_train.drop(['Cabin', 'Name', 'PassengerId'], axis=1)
df_test = df_test.drop(['Cabin', 'Name', 'PassengerId'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
X_train = df_train.drop('Transported', axis=1)
y_train = df_train['Transported']
X_test = df_test
```

```python
for i in X_train.columns:
    if X_train[i].dtype == object:
        X_train[i] = label.fit_transform(X_train[i])
```

```python
for i in X_test.columns:
    if X_test[i].dtype == object:
        X_test[i] = label.fit_transform(X_test[i])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
scaler = StandardScaler()
```

```python
model = RandomForestClassifier()
```

```python
pipe = Pipeline([('scaler', scaler), ('model',model)])
```

```python
params = {
    'model__criterion':["gini", "entropy", "log_loss"],
    'model__n_estimators':[60,80,100,120],
    'model__max_features':["sqrt", "log2", None],
    'model__bootstrap':[True,False],
    'model__oob_score':[True,False],
    'model__max_depth':[3,5,10,None]
    
}
```

```python
full_model = GridSearchCV(pipe,param_grid=params,cv=3,verbose=2,scoring='accuracy')
```

```python
full_model.fit(X_train,y_train)
```

```python
full_model.best_params_
```

```python
y_pred = full_model.predict(X_test)
```
